```python
x = np.random.randint(1,10,100)
y = np.random.randint(1,10,100)

m, b = np.polyfit(x, y, 1)  # 1 for linear regression

# Plot the scatter plot
plt.scatter(x, y)

# Generate x-values for the regression line (adjust range as needed)
x_reg = np.linspace(min(x), max(x), 100)  # 100 points for smooth line

# Calculate y-values for the regression line
y_reg = m * x_reg + b

# Plot the regression line
plt.plot(x_reg, y_reg, color='red')

# Add labels and title
plt.xlabel("Independent Variable")
plt.ylabel("Dependent Variable")
plt.title("Linear Regression")

plt.figure(figsize=(8, 5))

# Display the plot
plt.show()
```
