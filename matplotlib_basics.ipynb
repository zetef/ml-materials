```python
# Legend is added to figure by plt.legend()
x = np.linspace(0,50,100)
y1 = np.sin(x)

plt.plot(x, y1, label = 'Y1')
plt.plot(x, y2, label = 'Y2')
plt.xlabel('This is X Label')
plt.ylabel('This is Y Label')
plt.legend(loc='lower left')
```

```python
plt.figure(figsize=(10,8))

plt.plot(x, y1, color='red')
plt.plot(x, y2, color='y')
plt.plot(x, y2*2, color='#00FF00')
plt.plot(x, y2*3, color='0.8')
```

```python
## Line Styles
## I will also add line width(lw) and marker size

plt.figure(figsize=(10,8))

plt.plot(x, y1, color='red', ls='solid', marker='+', lw=3, markersize=3) 
plt.plot(x, y2, color='b', ls='dotted', marker='D', lw=4, markersize=4) 
plt.plot(x, y2*2, color='#00FF00', ls='dashed', marker='*', lw=5, markersize=5) 
plt.plot(x, y2*3, color='0.8', ls=':', marker='s', lw=6, markersize=6)
```

```python
x = np.linspace(0,100,1000)
y = np.square(x)

plt.plot(x,y)

plt.text(2,60, 'x=2, y=60', color='r')
```

```python
x = np.linspace(1,10,100)
y = np.square(x)

plt.plot(x, y)

plt.annotate('min',
             xy=(1,1),
             xytext=(1.5, 30), arrowprops=dict(facecolor='red'))

plt.annotate('max',
             xy=(10,100),
             xytext=(1,55),
             bbox=dict(boxstyle='round'),
             arrowprops=dict(arrowstyle='fancy', connectionstyle='angle3, angleA=0, angleB=-90'))

plt.annotate('mid', # string to annotate
             xy=(6,6**2), # the position of x and y
             xytext=(5, 12), # the xy position to add the texts
             bbox=dict(boxstyle='round', fc='none'), # or boxstyle can be 'round'
             arrowprops=dict(arrowstyle='wedge, tail_width=0.5', alpha=0.3))
```

```python
fig = plt.figure()

ax = plt.axes()

ax.plot(x, y, 'g')

ax.set_title('title')

ax.set_xlabel('X')
ax.set_ylabel('Y')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'g')
axes[1].plot(x,y,'b')

axes[0].set_title('f')
axes[1].set_title('g')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3)

for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('f')

plt.tight_layout()
```

```python
ax = plt.axes()

ax.plot(x,y,'g')
ax.set(title='f',
       xlabel='x',
       ylabel='y',
       ylim=(-10,10),
       xlim=(0,25))
```
